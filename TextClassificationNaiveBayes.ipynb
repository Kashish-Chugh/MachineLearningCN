```python
def fit(feature_set,labels):
    result={}   #initialising the dictionary to save all the counts
    labels_value=set(labels)     #getting ditinct label in the labels list
    #print(labels_value)
    for label in labels_value:
        current_row=[]
        #going through each distinct label
        result[label]={}            #to save each row with label as label 
        current_class_rows=(labels==label)        #this will give each row with label as label
        #print(current_class_rows)
        feature_current=feature_set[current_class_rows]   #create a feature set with only rows where label is label
        labels_current=labels[current_class_rows]     #create a output set with only the current label
        #print(feature_current)
        num_features=len(feature_set[0])
        result[label]["total"]=0      #total features with current label (initialised to 0)
        for j in range(0,num_features):
            #updating the sum when a word is found
            result[label][j]= feature_current[:,j].sum()
        for j in range(0,num_features):
            #add all the rows with current label
            result[label]["total"]+=result[label][j]
    return result
```

```python
def probability(count,x,current_class):
    #to find the probability that a given document is of given class
    num_features=len(count[current_class].keys())-1   #total count
    prob=0.0        #initialising the probability as we have to add log of probability of each word
    for j in range(len(x)):
        if(x[j]!=0):
            output=np.log((count[current_class][j])+1)-np.log((count[current_class]["total"])+num_features)
            prob+=output
    return prob
```

```python
def predictionSinglePoint(count,x):
    #to find the pedinction of the given document
    classes=count.keys()
    best_p=-99999
    best_c=""
    first_run=True
    for current_class in classes:
        #checking for each label and returning the one with highest probability
        if current_class=="total_data":     
            continue
        p_current_class=probability(count,x,current_class)
        if(first_run or p_current_class>best_p):
            best_p=p_current_class
            best_c=current_class
        first_run=False
    return best_c
```

```python
def predict(count,x_test):
    #function to predict the label of all the data
    y_pred=[]
    for x in x_test:
        x_class=predictionSinglePoint(count,x)
        y_pred.append(x_class)
    return y_pred
```

```python
#fitting the data and save the counts in count
count=fit(x_train,y_train)
```

```python
y_pred_own=predict(count,x_test)
```
